# Preprocessing Data #

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
features_one_hot = pd.get_dummies(data = df, columns = ['manufact', 'model'], drop_first= True)
```

```python
features_one_hot = features_one_hot.astype('float64', errors = 'ignore')
```

```python
features_one_hot.head()
```

```python
Y = features_one_hot['price'].to_numpy()
X = features_one_hot.drop(columns = ['sales', 'type', 'price', 'partition']).to_numpy()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X.astype('float64'))
```

```python
X.shape, Y.shape
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.25, random_state = 42)
xtrain.shape, xtest.shape
```
