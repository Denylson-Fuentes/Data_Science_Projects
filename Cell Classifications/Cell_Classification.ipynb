# Preprocessing Data #

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df['Class'] = df['Class'].replace([2, 4], [0, 1])
```

```python
df['BareNuc'] = df['BareNuc'].replace(['?'], np.nan)
```

```python
df['BareNuc'] = df['BareNuc'].astype(np.float64)
```

```python
df['BareNuc'] = df['BareNuc'].replace([np.nan], float(int(df['BareNuc'].mean())))
```

```python
df.head()
```

```python
features_one_hot = pd.get_dummies(data = df, columns = ['Clump', 'UnifSize', 'UnifShape', 'MargAdh', 'SingEpiSize', 'BareNuc', 'BlandChrom', 'NormNucl', 'Mit'], drop_first = True)
```

```python
features_one_hot = features_one_hot.astype('float64', errors = 'ignore')
```

```python
features_one_hot.head()
```

```python
x = features_one_hot.iloc[:, 1:]
y = features_one_hot.iloc[:, 0]
x.shape, y.shape, features_one_hot.shape
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x.astype('float64'))
```

```python
from numpy.random.mtrand import random
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape, x_test.shape
```
